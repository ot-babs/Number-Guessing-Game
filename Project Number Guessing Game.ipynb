```python
import random
#Define Global Variables
lower = 1
higher = 10
max_tries = 3

## Step 1: Define the function to play the game and initialise some useful values
def play_game():
    print("Guess a number between 1 and 10, in 3 tries. Type 500 to quit the game")
    rand_number = random.randint(lower, higher) # generates a random number betweeen 1 and 10 from our global value
    guess, tries = 0, 0 #Define the variable for guesses and tries to 0 
    trys = max_tries 
    

    # Core of our game: While loop used to check if the users entered guess is the same is the as the random number generated
    while guess != rand_number:
        try:
            guess = int(input("Take a guess: "))

            if guess == 500:
                print("You have decided to quit the game. Thanks for playing")
                break

            if guess > rand_number: # check if your guessed number is greater than the correct number
                print(f"Your Guess {guess} is Greater than the random number ")
                print(f"You have {trys-1} try{'s' if trys > 2 else ''} left.") 

            if guess < rand_number: # check if your guessed number is less than the correct number
                print (f"Your Guess {guess} is lower than the random number")
                print(f"You have {trys-1} try{'s' if trys > 2 else ''} left.") 
            
           
            tries += 1 #adds to the number of tries the the user as entered
            trys -= 1 #removes 1 from our maximum number of tries 
       
        except ValueError: print('Incorrect input, Please enter an integer value')

        if tries == max_tries:    # Checks if the user has tried the same amount as the games max tries 
            print("Unlucky! You have lost the game. You are out of guesses.")
            print(f"The correct answer was {rand_number}.")
            print("Please follow the instructions if you wish to play again or quit.")
            break


        if guess == rand_number: # check if you guessed the correct randomly generated numbebr 
            print(f"-> Hurray! The correct answer is {rand_number}. You got it in {tries} try{'s' if guess > 1 else ''}!")
            print("Please follow the instructions if you wish to play again or quit.")

        
 
 ## Step 2: Define the function Menu for the game. 
 ## Start the Function allows the user to start, restart quit the game
def menu():
   while True:
       print(' Press (1) to play the number game.')
       print('Press (Q) to play Quit the game.')
       choice = input('Please Enter 1 or Q to continue: ').lower()
       if choice == '1':
           play_game()
       elif choice == 'q':
            print("Goodbye")
            return False
       else: # condition is used to check if the user entered the correct input
           print(f'Not a correct choiceP: {choice} please try again') 

## Step 3: Start of Game
if __name__ == '__main__': 
    menu()
```
